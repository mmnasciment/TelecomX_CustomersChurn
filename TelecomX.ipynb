# ETL

Carregando os dados

```python
import numpy as np
import pandas as pd
import plotly.express as px
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
dados = pd.read_json(url)
dados.head()
```

Visualizando as colunas e seus tipos

```python
dados.info()
```

```python
list(dados.columns)
```

Desaninhando o JSON

```python
# Lista das colunas com dados aninhados
colunas_aninhadas = ['customer', 'phone', 'internet', 'account']
colunas_desaninhadas = ['customerID', 'Churn']
dados_desaninhados = pd.DataFrame()

for coluna in colunas_desaninhadas:
    dados_desaninhados[coluna] = dados[coluna]

# iteração sob a lista de colunas para normalizar os dados, com concatenação de cada nova coluna ao df com eixo = 1 para colunas
for coluna in colunas_aninhadas:
    normalizado = pd.json_normalize(dados[coluna])
    dados_desaninhados = pd.concat([dados_desaninhados, normalizado], axis=1)


dados_desaninhados

```

Verificando os tipos das novas colunas

```python
dados_desaninhados.info()
```

Transformando SenionCitizen de int64 para bool

```python
dados_desaninhados['SeniorCitizen'] = dados_desaninhados['SeniorCitizen'].astype(bool)
dados_desaninhados
```

Alterando nome de algumas colunas para facilitar querys. Eu já havia alterado algumas no decorrer do notebook, mas voltei e decidi alterar todas colunas aqui antes de começar a atratar os dados_desaninhados para ficar mais organizado

```python
dados_desaninhados.rename(columns={'seniorcitizen': 'is_senior_citizen'}, inplace=True)
dados_desaninhados.rename(columns={'phoneservice': 'phone_service'}, inplace=True)
dados_desaninhados.rename(columns={'multiplelines': 'multiple_lines'}, inplace=True)
dados_desaninhados.rename(columns={'internetservice': 'internet_service'}, inplace=True)
dados_desaninhados.rename(columns={'onlinesecurity': 'online_security'}, inplace=True)
dados_desaninhados.rename(columns={'onlinebackup': 'online_backup'}, inplace=True)
dados_desaninhados.rename(columns={'deviceprotection': 'device_protection'}, inplace=True)
dados_desaninhados.rename(columns={'techsupport': 'tech_support'}, inplace=True)
dados_desaninhados.rename(columns={'streamingtv': 'streaming_tv'}, inplace=True)
dados_desaninhados.rename(columns={'streamingmovies': 'streaming_movies'}, inplace=True)
dados_desaninhados.rename(columns={'paperlessbilling': 'paperless_billing'}, inplace=True)
dados_desaninhados.rename(columns={'paymentmethod': 'payment_method'}, inplace=True)
dados_desaninhados.rename(columns={'charges.monthly': 'monthly_charges'}, inplace=True)
dados_desaninhados.rename(columns={'charges.total': 'total_charges'}, inplace=True)
dados_desaninhados.rename(columns={'customerid': 'customer_id'}, inplace=True)

dados_desaninhados.head()
```

# Padronização e Transformação de Dados

Novo DF dados_padronizados

```python
url = 'https://raw.githubusercontent.com/mmnasciment/TelecomX_CustomersChurn/refs/heads/main/dados_tratados.csv'
dados_padronizados = pd.read_csv(url)
dados_padronizados
```

Padronização dos registros (garantir que todas colunas com "resposta" binária sigam o padrão binário Yes=1, No=0)

```python
dados_padronizados.info()
```

iterando sobre os valores de colunas tipo object

```python
for coluna in dados_padronizados.columns:
  if coluna != 'customer_id' and coluna != 'tenure' and coluna != 'monthly_charges' and coluna != 'total_charges' and coluna != 'daily_charges':
    print(dados_padronizados[coluna].value_counts())
```

Será utilizado o metodo 1-de-c para padronizar as colunas para criar um arquivo com formato binário

```python
lista_col_binaria = []
lista_col_nao_binaria = []

for coluna in dados_padronizados.columns:
  if coluna == 'customer_id':
    continue

  qntd_valores_unicos = dados_padronizados[coluna].nunique()
  if qntd_valores_unicos == 2:
    lista_col_binaria.append(coluna)
  else:
    lista_col_nao_binaria.append(coluna)

print(lista_col_binaria)
print(lista_col_nao_binaria)
```

Convertendo as colunas tipo object binarios (Yes e Male = 1, No e Female = 0)


```python
dados_binarios = dados_padronizados.copy()
for coluna in lista_col_binaria:
  dados_binarios[coluna] = dados_padronizados[coluna].map({'Yes': 1, 'No': 0, 'Male': 1, 'Female': 0})
```

Convertendo as colunas tipo object binarios com 3 registros (No =0, Yes = 1, Else = 2)

```python
#lista_col_nao_binaria.remove('daily_charges')
#lista_col_nao_binaria.remove('tenure')
#lista_col_nao_binaria.remove('monthly_charges')
#lista_col_nao_binaria.remove('total_charges')
#lista_col_nao_binaria


for coluna in lista_col_nao_binaria:
  if coluna != 'payment_method' and coluna != 'internet_service' and coluna != 'contract':
    dados_binarios[coluna] = dados_padronizados[coluna].map({'Yes': 1, 'No': 0, 'No internet service': 2, 'No phone service': 2})
```

# Convertendo as colunas tipo object não binários

contract ('Month-to-month': 0, 'One year': 1, 'Two year': 2)

```python
dados_binarios['contract'] = dados_padronizados['contract'].map({'Month-to-month': 0, 'One year': 1, 'Two year': 2})
```

internet_service ('Fiber optic': 2, 'DSL': 0, 'No': 1)

```python
dados_binarios['internet_service'] = dados_padronizados['internet_service'].map({'Fiber optic': 2, 'DSL': 1, 'No': 0})
```

payment_method ('Electronic check': 0, 'Mailed check': 1, 'Bank transfer (automatic)': 2, 'Credit card (automatic)': 3

```python
dados_binarios['payment_method'] = dados_padronizados['payment_method'].map({'Electronic check': 0, 'Mailed check': 1, 'Bank transfer (automatic)': 2, 'Credit card (automatic)': 3})
```

```python
for coluna in dados_binarios.columns:
  if coluna != 'customer_id' and coluna != 'tenure' and coluna != 'monthly_charges' and coluna != 'total_charges' and coluna != 'daily_charges':
    print(dados_binarios[coluna].value_counts())
```

Salvando dados_binarios

```python
dados_binarios.to_csv('dados_binarios.csv', index=False)
```
