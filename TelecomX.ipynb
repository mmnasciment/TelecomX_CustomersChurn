Carregando os dados

```python
import numpy as np
import pandas as pd
import plotly.express as px
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
dados = pd.read_json(url)
dados.head()
```

Visualizando as colunas e seus tipos

```python
dados.info()
```

```python
list(dados.columns)
```

Desaninhando o JSON

```python
# Lista das colunas com dados aninhados
colunas_aninhadas = ['customer', 'phone', 'internet', 'account']
colunas_desaninhadas = ['customerID', 'Churn']
dados_desaninhados = pd.DataFrame()

for coluna in colunas_desaninhadas:
    dados_desaninhados[coluna] = dados[coluna]

# iteração sob a lista de colunas para normalizar os dados, com concatenação de cada nova coluna ao df com eixo = 1 para colunas
for coluna in colunas_aninhadas:
    normalizado = pd.json_normalize(dados[coluna])
    dados_desaninhados = pd.concat([dados_desaninhados, normalizado], axis=1)


dados_desaninhados

```

Verificando os tipos das novas colunas

```python
dados_desaninhados.info()
```

Transformando SenionCitizen de int64 para bool

```python
dados_desaninhados['SeniorCitizen'] = dados_desaninhados['SeniorCitizen'].astype(bool)
dados_desaninhados
```

Alterando nome de algumas colunas para facilitar querys. Eu já havia alterado algumas no decorrer do notebook, mas voltei e decidi alterar todas colunas aqui antes de começar a atratar os dados_desaninhados para ficar mais organizado

```python
dados_desaninhados.rename(columns={'seniorcitizen': 'is_senior_citizen'}, inplace=True)
dados_desaninhados.rename(columns={'phoneservice': 'phone_service'}, inplace=True)
dados_desaninhados.rename(columns={'multiplelines': 'multiple_lines'}, inplace=True)
dados_desaninhados.rename(columns={'internetservice': 'internet_service'}, inplace=True)
dados_desaninhados.rename(columns={'onlinesecurity': 'online_security'}, inplace=True)
dados_desaninhados.rename(columns={'onlinebackup': 'online_backup'}, inplace=True)
dados_desaninhados.rename(columns={'deviceprotection': 'device_protection'}, inplace=True)
dados_desaninhados.rename(columns={'techsupport': 'tech_support'}, inplace=True)
dados_desaninhados.rename(columns={'streamingtv': 'streaming_tv'}, inplace=True)
dados_desaninhados.rename(columns={'streamingmovies': 'streaming_movies'}, inplace=True)
dados_desaninhados.rename(columns={'paperlessbilling': 'paperless_billing'}, inplace=True)
dados_desaninhados.rename(columns={'paymentmethod': 'payment_method'}, inplace=True)
dados_desaninhados.rename(columns={'charges.monthly': 'monthly_charges'}, inplace=True)
dados_desaninhados.rename(columns={'charges.total': 'total_charges'}, inplace=True)
dados_desaninhados.rename(columns={'customerid': 'customer_id'}, inplace=True)

dados_desaninhados.head()
```

Padronizando nome das colunas

```python
dados_desaninhados.columns = dados_desaninhados.columns.str.lower()
dados_desaninhados.head()
```

Convertendo TotalCharges para float64 (Erro pois existem campos com valores incoerentes)
Campos retirados e conversão feita

```python
dados_desaninhados['total_charges'] = dados_desaninhados['total_charges'].astype(float)
dados_desaninhados.info()
```

Identificando e quantificando quais registros possuem inconsistencias

```python
dados_desaninhados[dados_desaninhados['total_charges'] == ' '].index
```

Retirando dados (menos 11 registros)

```python
dados_desaninhados = dados_desaninhados.drop(dados_desaninhados[dados_desaninhados['total_charges'] == ' '].index)
dados_desaninhados.info()
```

# Procurando inconsistencias

Dados ausentes(NA) não existem

```python

for coluna in dados_desaninhados.columns:
    print(f'{coluna}: {dados_desaninhados[coluna].isna().sum()}')
```

Dados vazios ('') existem na coluna Churn, com 224 ocorrencias. Os dados podem ser excluidos ou preenchidos aleatoriamente com base na média/moda. Escolhi excluir os dados por representarem apenas 3% da amostra. Dessa forma, os registros mantidos são confiáveis e fidedignos.

```python
for coluna in dados_desaninhados.columns:
  indice_coluna = dados_desaninhados[dados_desaninhados[coluna] == ''].index
  print(f' coluna {coluna}: indice{indice_coluna}')
```

```python
dados_desaninhados = dados_desaninhados.drop(dados_desaninhados[dados_desaninhados['churn'] == ''].index)
dados_desaninhados.info()
```

Verificando os registros existentes em cada coluna em busca de inconsistencias

Padronizando is_senior_citizen para padrao da tabela(Yes, No)

# Coluna de Contas diárias

Criando novo DF dados_tratados

```python
dados_tratados = dados_desaninhados.copy()
dados_tratados.head()
```

Criando nova coluna e assumindo os meses com 30 dias para facilitar a conta

```python
dados_tratados['daily_charges'] = dados_tratados['monthly_charges'] / 30.0
dados_tratados.daily_charges
```

arredondando as casas para padronização

```python
dados_tratados['daily_charges'] = dados_tratados['daily_charges'].round(2)
dados_tratados.head()
```
